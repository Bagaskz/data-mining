**6. Untuk menghapus data duplicate**

```python
df = df.drop_duplicates()
```

**7. Untuk menghapus data yang hilang dari kolom B-10 sampai B-14(missing values)**

```python
df['B-10 (Kg)'] = df['B-10 (Kg)'].fillna(df['B-10 (Kg)'].mean())
df['N-10 (USD)'] = df['N-10 (USD)'].fillna(df['N-10 (USD)'].mean())
df['B-11 (Kg)'] = df['B-11 (Kg)'].fillna(df['B-11 (Kg)'].mean())
df['N-11 (USD)'] = df['N-11 (USD)'].fillna(df['N-11 (USD)'].mean())
df['B-12 (Kg)'] = df['B-12 (Kg)'].fillna(df['B-12 (Kg)'].mean())
df['N-12 (USD)'] = df['N-12 (USD)'].fillna(df['N-12 (USD)'].mean())
df['B-13 (Kg)'] = df['B-13 (Kg)'].fillna(df['B-13 (Kg)'].mean())
df['N-13 (USD)'] = df['N-13 (USD)'].fillna(df['N-13 (USD)'].mean())
df['B-14 (Kg)'] = df['B-14 (Kg)'].fillna(df['B-14 (Kg)'].mean())

```

**8. Menampilkan data yang sudah di cleansing**

```python
print(df)
```
